```python
#-*- coding:utf-8 -*-
from urllib.error import HTTPError
from selenium import webdriver

file_path='C:\\Users\\user\\PycharmProjects\\'
file_name='newsId.txt'

newsurl='https://news.naver.com/main/ranking/popularDay.nhn?rankingType=popular_day&sectionId='
#sectionId 만 변경
#urlid='https://news.naver.com/main/ranking/popularDay.nhn?rankingType=popular_day&sectionId=105&date=20200729'


def make_request():
    p_file = open(file_path + file_name,'r') #읽기전용 파일 로딩
    
    for p_code in p_file.readlines():
        try:
            url = newsurl+p_code
            driver = webdriver.Chrome('C:\\Users\\user\\Downloads\\chromedriver.exe')
            driver.get(url) 
            print(url)
      #      search_result = driver.find_element_by_xpath("/html/body/div/table/tbody/tr/td[2]/div/div[3]/div/ul/li[2]/a")#xpath로 클릭할 오브젝트 얻어냄
            search_result = driver.find_element_by_xpath("//*[@id='wrap']/table/tbody/tr/td[2]/div/div[3]/div/ul/li[2]/a")
            search_result.click()
            crawling(driver)
            print('search_result : ',search_result)
        except HTTPError as e:
            pass
        
    p_file.close()    
    
    
def crawling(driver):
    crawling_results = []    
    #
    crawling_results.append(driver.find_elements_by_tag_name('h4')[0].text) #title
    print('crawling_results',crawling_results)

    
make_request()
```
